# Contornar problemas previstos no sistema

Desenvolva um programa que recebe do usuário nome completo e ano de nascimento que seja entre 1922 e 2021.
<br></br>
A partir dessas informações, o sistema mostrará o nome do usuário e a idade que completou, ou completará, no ano atual (2022).
<br></br>
Caso o usuário não digite um número ou apareça um inválido no campo do ano, o sistema informará o erro e continuará perguntando até que um valor correto seja preenchido.

```python
from datetime import date

def digitarNome():
  print("Olá, digite seu nome completo:")
  while True:
    nomeUsuario = str(input())
    if nomeUsuario != '':
      break
  return nomeUsuario

def digitarIdade():
  while True:
    nomeUsuario = digitarNome()
    print("Olá, digite seu ano de nascimento:")
    anoNascimentoUsuario = int(input())
    if(anoNascimentoUsuario < 1922 or anoNascimentoUsuario > 2024):
      print("Favor informar data de nascimento acima de 1922 e menor que 2024")
      continue

    idadeAtualUsuario = date.today().year - anoNascimentoUsuario
    print(nomeUsuario, "você completará", idadeAtualUsuario, "anos em 2024")
    return

def contonarProblemasSistema():
  digitarIdade()
```

```python
contonarProblemasSistema()
```
